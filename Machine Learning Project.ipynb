### Dropping Columns which are not required 

```python
dataset = dataset.drop(['App'],axis=1)
dataset = dataset.drop(['Genres'],axis=1)
dataset = dataset.drop(['Last Updated'],axis=1)
```

### Filling missing Numerical Values (If any) by mean of each column 

```python
dataset.fillna(dataset.mean(), inplace=True)
```

### Rounding off numerical values to two decimal points

```python
dataset.round(2)
```

```python
dataset.shape
```

### Getting the label to apply classification

```python
label = dataset['Category']
```

```python
dataset = dataset.drop(['Category'],axis=1)
```

### Checking for any Null Values

```python
dataset.isnull().any()
```

### Filling any Strings with Mode of Value (If any)

```python
dataset.fillna(dataset.mode, inplace=True)
```

### Label Encoding
### Change values to numerical for our model to understand

```python
from sklearn.preprocessing import LabelEncoder
dataset = dataset.apply(LabelEncoder().fit_transform)
```

### Spliting dataset to Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset, label, test_size = 0.2, random_state = 0)
```

### Importing Models from SKLEARN
### 1) K-nearest neighbors
### 2) Support Vector Classification
### 3) Decision Tree
### 4) Random Forst 
### 5) AdaBoost classifier
### 6) Gradient Boosting Classification
### 7) Gaussian Naive Bayes
### 8) Bernoulli Naive Bayes
### 9) Multinomial Naive Bayes 
### 10) LinearDiscriminantAnalysis

```python
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

### Loading Objects of classifiers into an Array

```python
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="rbf", C=0.025, probability=True),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    BernoulliNB(),
    MultinomialNB(),
    LinearDiscriminantAnalysis()]
```

```python
log_cols=["Classifier", "Accuracy"]
log = pd.DataFrame(columns=log_cols)

for clf in classifiers:
    clf.fit(X_train, y_train)
    name = clf.__class__.__name__
    print("="*50)
    print(name)
    
    print('Predicted values')
    train_predictions = clf.predict(X_test)
    print(train_predictions)
    acc = accuracy_score(y_test, train_predictions)
    print("Accuracy: {:.4%}".format(acc))
    log_entry = pd.DataFrame([[name, acc * 100]], columns=log_cols)
    log = log.append(log_entry)

print("="*50)


 
```
